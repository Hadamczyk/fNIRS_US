```python
# Importowanie niezbędnych modułów
import mne
from mne.io import read_raw_snirf
```

Co robi ta komórka?

```python
# Ścieżka do pliku SNIRF
# Możesz użyć narzędzia do uploadowania plików w Google Colab, aby przesłać plik SNIRF
from google.colab import files
uploaded = files.upload()

# Po uploadzie plik będzie dostępny w zmiennej uploaded
file_path = list(uploaded.keys())[0]

# Wczytanie danych
raw = read_raw_snirf(file_path, preload=True)

# Wyświetlenie podstawowych informacji o danych
print(raw.info)

# Wyświetlenie sygnału
raw.plot()

# Wyświetlenie konfiguracji montażu (opcjonalne)
raw.plot_sensors()

# Wyświetlenie informacji o czasach zdarzeń (opcjonalne)
events, event_dict = mne.events_from_annotations(raw)
mne.viz.plot_events(events, event_id=event_dict)
```
